```python
import pandas as pd
from database import database
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
```

```python
# Pulling data from database
db = database()
df = pd.DataFrame(db.select("player_all_stats_cleaned"))
db.close()
```

```python
# Things to consider look into:

# player characteristics vs positions
# position distribution over the years
# game participation vs positions
# shooting stats vs positions
```

```python
# player characteristics vs position

fig = go.Figure()
fig.add_trace(go.Box(x=df.pos, y=df.height, visible=True))

buttons = []
# Add buttons for each variable
for col in ['height', 'weight', 'age']:
    unit = 'cm' if col == 'height' else ('lbs' if col == 'weight' else 'years')
    data = {'x':[df.pos],
             'y':[df[col]],
             'type':'box',}
    layout = {'title': f"{col.title()} vs Positions",
              'xaxis': {'title': 'Positions'},
              'yaxis': {'title': f"{col.title()} ({unit})"}}
    buttons.append(dict(method='update',
                        label=col,
                        visible=True,
                        args=[data, layout, [0]]))

updatemenus = list([
    dict(buttons=buttons,
         direction='down',
         showactive=True,
         x=0,
         y=1.15,
         xanchor="center",
         yanchor="top")
])

# add dropdown menus to the figure
fig.update_layout(updatemenus=updatemenus, title={
        'text': "Height vs Positions",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',},
        xaxis={'title': 'Positions'},
        yaxis={'title': 'Height (cm)'})
fig.show()
```

```python
# Player position distribution for each year
traces = []
positions = df["pos"].unique()
for pos in positions:
    temp_df = df[df["pos"] == pos]
    temp_yr = temp_df["year"].value_counts()
    traces.append(go.Bar(x=temp_yr.index, y=temp_yr.values, name=str(pos)))
layout = go.Layout(
    barmode='group',
    title={
        'text': "Positions distribution over last five year",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',},
    xaxis={'title': 'Positions'},
    yaxis={'title': 'Number of Players'}
)
fig = go.Figure(data=traces, layout=layout)
fig.show()
```

```python
# Games and games started vs positions

games_df = df.pivot_table(index="pos", values=["g", "gs"], aggfunc=np.sum)

fig = go.Figure()
traces = []
traces.append(go.Bar(x=games_df.index, y=games_df.g, name='Games Played'))
traces.append(go.Bar(x=games_df.index, y=games_df.gs, name='Games Started'))
layout = go.Layout(
    barmode='group',
    title={
        'text': "Total games played/started vs Positions ",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis={'title': 'Positions'},
    yaxis={'title': 'Games'}
)
fig = go.Figure(data=traces, layout=layout)
fig.show()
```
